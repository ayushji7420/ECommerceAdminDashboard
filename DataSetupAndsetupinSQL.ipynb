```python
import pandas as pd
import numpy as np
import random as rd
from faker import Faker
import bcrypt
fake = Faker('en_IN')
```

```python
#-----order_item_id-----
order_item_Id = list(range(1,541910))
df["order_item_id"] = order_item_Id
df
```

```python
# -----order_id-----
order_Id = df["InvoiceNo"].unique()
order_Id
```

```python
#-----User_ID-----
user_Id = df["CustomerID"].unique()
user_Id = user_Id.astype(np.int64)
user_Id
```

```python
#-----Product_Id-----
product_Id = df["StockCode"].unique()
product_Id
```

# Product Details Data

```python
#Product_ID vs price dictionary
product_vs_priceperunit = []
for a in product_Id:
    price = float(df.loc[df["StockCode"] == a, "UnitPrice"].values[0])
    val = str(df.loc[df["StockCode"] == a, "Description"].values[0])
    quant = int(rd.randint(200,300))
    product_vs_priceperunit.append({
        "product_Id": a,
        "UnitPrice": price,
        "Product_Description": val,
        "Quantity": quant
    })

product_vs_priceperunit
```

```python

product_df = pd.DataFrame(product_vs_priceperunit, columns=["product_Id", "UnitPrice", "Product_Description", "Quantity"])
category = ["Clothes", "Electronics", "Decor"]
category_lst = []
for _ in product_Id:
    cat = rd.choice(category)
    category_lst.append(cat)
product_df["Category"] = category_lst
product_df
```

```python
product_df.columns = ["product_id", "unitprice", "description", "stock_quantity", "category"]
product_df.sort_values(by="product_id", ascending=True)
```

# Order Data

```python
pay_stats = ["Paid", "Unpaid"]
delivery_stats = ["Packaging", "Delivered", "In-Transit"]

df["Amount"] = df["Quantity"]*df["UnitPrice"]
order_data = []
for oid in order_Id:
    Amount = df.loc[df["InvoiceNo"] == oid, "Amount"]
    total_amount = float(sum(Amount))
    user_no = df.loc[df["InvoiceNo"] == oid, "CustomerID"].values[0]
    order_data.append({
        'order_Id' : oid,
        'user_Id': user_no,
        'Total_Amount': total_amount,
        'Delivery_Status': rd.choice(delivery_stats),
        'Payment_Status': rd.choice(pay_stats)
    })

order_df = pd.DataFrame(order_data)
```

```python
order_date = []
for oid in order_Id:
    date = str(df.loc[df["InvoiceNo"]==oid, "InvoiceDate"].values[0])
    order_date.append(date)

order_df["order_date"] = order_date
order_df
```

```python
order_df.columns = ["order_id", "user_id", "total_amount", "delivery_status", "payement_status", "order_date"]
order_df
```

```python
order_df['order_date'] = pd.to_datetime(order_df['order_date'], dayfirst=False)
```

```python
print(order_df["user_id"].isnull().sum())
```

```python
order_df = order_df.dropna(subset=['user_id'])
order_df
```

# User Data

```python
user_data = []

for uid in user_Id:
    password = fake.password()
    hashed_pass = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())
    user_data.append({
        'user_Id': uid,
        'username': fake.name(),
        'contact': fake.phone_number(),
        'password': hashed_pass,
        'location': fake.city()
        
    })

user_df = pd.DataFrame(user_data)
user_df
```
